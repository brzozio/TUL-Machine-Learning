```python
import pandas as pd
import os
from json import dump

base_path : str = os.path.dirname(os.getcwd())
csv_path  : str = base_path + '\\csv'
src_path  : str = base_path + '\\src'
json_path : str = base_path + '\\json'

num_of_movies = 200

def distance_set(list1, list2) -> float:
    return 1.0 - len(set(list1)&set(list2))/len(set(list1)|set(list2))

movies = pd.read_csv(csv_path + "\\MOVIES_DATA.csv", sep=',').drop('Unnamed: 0', axis=1)
```

```python
movies
```

```python
distance_movies = []
for movie1 in movies["popularity"]:
    for movie2 in movies["popularity"]:
        distance_movies.append([abs(movie1 - movie2)])

for id1, movie1 in enumerate(movies["rating"]):
    for id2, movie2 in enumerate(movies["rating"]):
        distance_movies[id1*len(movies["rating"])+id2].append(abs(movie1 - movie2))


for id1, movie1 in enumerate(movies["director"]):
    for id2, movie2 in enumerate(movies["director"]):
        distance_movies[id1*len(movies["director"])+id2].append(distance_set(eval(movie1),eval(movie2)))


for id1, movie1 in enumerate(movies["actors"]):
    for id2, movie2 in enumerate(movies["actors"]):
        distance_movies[id1*len(movies["actors"])+id2].append(distance_set(eval(movie1),eval(movie2)))

for id1, movie1 in enumerate(movies["genres"]):
    for id2, movie2 in enumerate(movies["genres"]):
        distance_movies[id1*len(movies["genres"])+id2].append(distance_set(eval(movie1),eval(movie2)))
```

```python
distance_popularity_max = distance_movies[0][0]
distance_popularity_min = distance_movies[0][0]
distance_rating_max = distance_movies[0][1]
distance_rating_min = distance_movies[0][1]
for movie_pair in distance_movies:
    if(distance_popularity_max < movie_pair[0]): distance_popularity_max = movie_pair[0]
    if(distance_popularity_min > movie_pair[0]): distance_popularity_max = movie_pair[0]
    if(distance_rating_max < movie_pair[1]): distance_rating_max = movie_pair[1]
    if(distance_rating_min > movie_pair[1]): distance_rating_min = movie_pair[1]

for movie_pair in distance_movies:
    movie_pair[0] = (movie_pair[0] - distance_popularity_min)/distance_popularity_max
    movie_pair[1] = (movie_pair[1] - distance_rating_min)/distance_rating_max
```

```python
from math import sin, cos

PI = 1.57079632679
ANGLE_RESOLUTION = 5

ANGLES = []
for angle in range(ANGLE_RESOLUTION):
 ANGLES.append(angle*PI/(ANGLE_RESOLUTION-1))

SINES = []
COSINES = []
for angle in ANGLES:
 SINES.append(sin(angle))
 COSINES.append(cos(angle))

SINES[0] = 0
SINES[ANGLE_RESOLUTION-1] = 1
COSINES[0] = 1
COSINES[ANGLE_RESOLUTION-1] = 0

PARAMS_WEIGHTS = []

for i in range(ANGLE_RESOLUTION):
    for j in range(ANGLE_RESOLUTION):
        for k in range(ANGLE_RESOLUTION):
            for l in range(ANGLE_RESOLUTION):
                PARAMS_WEIGHTS.append([SINES[i], COSINES[i]*SINES[j], COSINES[i]*COSINES[j]*SINES[k], COSINES[i]*COSINES[j]*COSINES[k]*SINES[l], COSINES[i]*COSINES[j]*COSINES[k]*COSINES[l]])

len(PARAMS_WEIGHTS)
```
